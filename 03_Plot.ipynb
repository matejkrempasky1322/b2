## Jednoduchý graf

```
import matplotlib.pyplot as plt

plt.plot([1, 2, 3])
plt.title('Jednoduchý graf')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Graf s parametrami

Príklad amplitúdovo modulovaného signálu:

```
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 1, 1000)
signal = np.sin(2 * np.pi * 10 * t) * (1 + 0.5 * np.sin(2 * np.pi * 1 * t))

plt.figure(figsize=(10, 4))
plt.plot(t, signal, color='#2ecc71')
plt.title('Amplitúdovo modulovaný signál')
plt.xlabel('Čas (s)')
plt.ylabel('Amplitúda')
plt.grid(True, alpha=0.3)
plt.show()
```
