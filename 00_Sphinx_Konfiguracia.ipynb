# Konfigurácia

Vlastnosti a vzhľad dokumentov vytváraných na platforme Sphinx môžeme konfigurovať rozšíreniami a grafickými témami. Rovnako môžeme doplniť a rozširovať aj vlastnosti formátovania textov v Markdown.

## Sphinx

Distribúcia Sphinx obsahuje základnú sadu [rozšírení](https://www.sphinx-doc.org/en/master/usage/extensions/index.html) pre formátovanie a konverziu textov. Ďalšie rozšírenia je možné inštalovať ako samostatné balíky. Externé rozšírenia sa inštalujú pomocou inštalátora pip a aktivujú sa v súbore `conf.py`.

## MyST-NB

Rozšírenie [MyST-NB](https://myst-nb.readthedocs.io/en/v0.13.2/index.html) umožňuje použitie zdrojových textov vo formáte Markdown (.md) a Jupyter Notebook (.ipynb) na platforme Sphinx.

Inštalácia:

```python
pip install myst-nb
```

Aktivácia v `conf.py`:

```python
extensions = [
    ...,
    "myst_nb"
]
```

## Sphinx-copybutton

Rozšírenie [Sphinx-copybutton](https://sphinx-copybutton.readthedocs.io/en/latest/index.html) pridá do buniek kódu ikonu pre skopírovanie obsahu do schránky.

Inštalácia:

```python
pip install sphinx-copybutton
```

Aktivácia v `conf.py`:

```python
extensions = [
    ...,
    "sphinx_copybutton"
]
```

## Sphinx-design

Rozšírenie ktoré podporuje konfigurovateľné [komponenty](https://sphinx-design.readthedocs.io/en/latest/index.html) pre témy webových rozhraní ako sú kaskádové štýly, expandovateľné kontajnery, tlačidlá, ikony a ďalšie.

Dostupné témy grafického rozhrania:

- [alabaster](https://alabaster.readthedocs.io/en/latest/)
- [sphinx-book-theme](https://sphinx-book-theme.readthedocs.io/en/stable/index.html)
- [pydata-sphinx-theme](https://pydata-sphinx-theme.readthedocs.io/en/stable/index.html)
- [sphinx-rtd-theme](https://sphinx-rtd-theme.readthedocs.io/en/stable/)
- [furo](https://pradyunsg.me/furo/)

Inštalácia:

```python
pip install sphinx-design
pip install sphinx-rtd-theme
```

Konfigurácia témy v `conf.py`:

```python
html_theme = 'sphinx_rtd_theme'
html_logo = "my_logo.png"
html_static_path = ['_static']
```

Ďalšie témy grafického rozhrania sú dostupné na [internete](https://sphinx-themes.org/).

## MyST

Pre formátovanie textov v [MyST](https://myst-parser.readthedocs.io/en/latest/syntax/optional.html) sú dostupné zabudované rozšírenia:

- `smartquotes` - zjednodušené zadávanie typografických značiek
- `strikethrough` - prečiarknutý text
- `dollarmath` - zadávanie matematických textov pomocou `$...$`
- `amsmath` - doplnenie prostredia amsmath pre LaTeX
- `colon_fence` - alternatívne použitie `:::` pre zobrazenie kódov
- `substitution` - definovanie substitučných reťazcov

Aktivácia v `conf.py`:

```python
myst_enable_extensions = [
    "amsmath",
    "colon_fence",
    "dollarmath",
    "smartquotes",
    "strikethrough",
    "substitution",
]
```
