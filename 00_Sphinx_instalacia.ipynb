# Inštalácia Sphinx

Sphinx je multiplatformová aplikácia vytvorená v jazyku Python. Inštalácia pre rôzne platformy je popísaná v [dokumentácii](https://www.sphinx-doc.org/en/master/usage/installation.html). Prehliadanie publikácií vytvorených pomocou Sphinx nevyžaduje okrem štandardného www prehliadača inštaláciu doplnkového programového vybavenia.

## Linux

Pre distribúciu Debian/Ubuntu môžeme nainštalovať globálnu verziu Sphinx pomocou inštalátora apt-get:

```
pip install sphinx sphinx-rtd-theme nbsphinx
```

alebo z [Python Package Index](https://pypi.org/project/Sphinx/) (PyPI) pomocou inštalátora pip:

````python
pip install -U sphinx
````

Druhý spôsob umožňuje paralelnú inštaláciu rôznych verzií Sphinx a ich konfigurácií vo virtuálnych prostrediach vytvorených pomocou [venv](https://docs.python.org/3/library/venv.html).

Aktuálnu verziu Sphinx získame pomocou príkazu:

```
sphinx-build --version
```

Nainštalovaná verzia Sphinx konvertuje texty vytvorené vo formáte reStructuredText. Pre tvorbu publikácií v značkovacom jazyku Markdown doinštalujeme prekladač MyST, popis inštalácie je v [dokumentácii](https://myst-parser.readthedocs.io/en/latest/intro.html):

```
pip install myst-parser
```

## Windows

Python nie je štandardnou súčasťou distribúcie Windows, pre jeho inštaláciu využite návody dostupné na internete, napríklad [The Hitchhiker's Guide to Python](https://docs.python-guide.org/). Doporučená verzia pre inštaláciu je Python 3. Po inštalácii nainštalujte Sphinx pomocou inštalátora pip.
