```python
# ============================================================
# PARAMETRE – zmeň tieto hodnoty a spusti bunku znova (Shift+Enter)
# ============================================================
m  = 1.0    # hmotnosť telesa [kg]         → skús napr. 0.5 alebo 2.0
k  = 10.0   # tuhosť pružiny [N/m]         → väčšie k = rýchlejšie kmitanie
b  = 1.0    # koeficient tlmenia [N·s/m]   → 0 = bez tlmenia, >6.3 = pretlmený
x0 = 1.0    # počiatočná výchylka [m]      → kde začína teleso
v0 = 0.0    # počiatočná rýchlosť [m/s]    → 0 = teleso štartuje zo spokoja
# ============================================================

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

%matplotlib inline
plt.style.use('dark_background')

# --- Výpočet základných veličín ---
b_krit = 2 * np.sqrt(k * m)   # kritické tlmenie – hranica medzi kmitaním a tlmením
omega0 = np.sqrt(k / m)        # vlastná uhlová frekvencia [rad/s]
pomer  = b / b_krit            # pomer tlmenia ζ (zeta)

# --- Určenie režimu tlmenia podľa pomeru ζ ---
if pomer < 0.99:
    rezim = 'PODTLMENÝ'        # teleso kmitá s útlmom
    farba_rezim = '#00FF88'
elif pomer <= 1.01:
    rezim = 'KRITICKY TLMENÝ'  # najrýchlejší návrat bez kmitania
    farba_rezim = '#FFD700'
else:
    rezim = 'PRETLMENÝ'        # teleso sa pomaly vracia bez kmitania
    farba_rezim = '#FF6B6B'

# --- Pohybová rovnica: m·x'' + b·x' + k·x = 0 ---
# Prepisujeme na sústavu dvoch rovníc prvého rádu:
# dx/dt = v
# dv/dt = (-b·v - k·x) / m
def pohybova_rovnica(t, y):
    x, v = y           # y[0] = výchylka, y[1] = rýchlosť
    dxdt = v
    dvdt = (-b * v - k * x) / m
    return [dxdt, dvdt]

# --- Časový rozsah simulácie ---
T_kyvov = 2 * np.pi / omega0           # perióda jedného kyvu [s]
t_max   = max(15 * T_kyvov, 5)         # simulujeme 15 periód
t_span  = (0, t_max)
t_eval  = np.linspace(0, t_max, 2000)  # 2000 časových krokov

# --- Numerické riešenie pomocou SciPy (metóda Runge-Kutta RK45) ---
sol = solve_ivp(
    pohybova_rovnica,
    t_span,
    [x0, v0],      # počiatočné podmienky [výchylka, rýchlosť]
    t_eval=t_eval,
    method='RK45',
    rtol=1e-8      # presnosť riešenia
)

# --- Výsledky zo solvera ---
t = sol.t      # časová os
x = sol.y[0]   # výchylka v čase
v = sol.y[1]   # rýchlosť v čase

# --- Obálka útlmu (len pre podtlmený režim) ---
# Obálka ukazuje ako exponenciálne klesá maximálna výchylka
if pomer < 1:
    alpha  = b / (2 * m)
    obalka = x0 * np.exp(-alpha * t)

# --- Vytvorenie grafu (4 podgrafy) ---
fig = plt.figure(figsize=(13, 10))
fig.patch.set_facecolor('#0d0d1a')

# Rozmiestnenie podgrafov: [ľavý okraj, dolný okraj, šírka, výška]
ax1 = fig.add_axes([0.07, 0.55, 0.55, 0.38])  # výchylka
ax2 = fig.add_axes([0.68, 0.55, 0.28, 0.38])  # fázový diagram
ax3 = fig.add_axes([0.07, 0.08, 0.55, 0.35])  # rýchlosť
ax4 = fig.add_axes([0.68, 0.08, 0.28, 0.35])  # porovnanie režimov

# ── Graf výchylky ─────────────────────────────────────────
ax1.set_facecolor('#0d0d1a')
ax1.plot(t, x, color='#00BFFF', linewidth=2.2, label='Výchylka x(t)')
ax1.axhline(y=0, color='white', linewidth=1, linestyle='--', alpha=0.3)

if pomer < 1:
    # Žltá čiarkovaná čiara ukazuje obálku útlmu
    ax1.plot(t,  obalka, color='#FFD700', linewidth=1.5,
             linestyle='--', alpha=0.7, label='Obálka útlmu')
    ax1.plot(t, -obalka, color='#FFD700', linewidth=1.5,
             linestyle='--', alpha=0.7)

ax1.fill_between(t, x, 0, where=(x >= 0), alpha=0.08, color='#00BFFF')
ax1.fill_between(t, x, 0, where=(x < 0),  alpha=0.08, color='#FF6B6B')
ax1.set_xlabel('Čas t (s)', color='#cccccc', fontsize=10)
ax1.set_ylabel('Výchylka x (m)', color='#cccccc', fontsize=10)
ax1.set_title('Závislosť výchylky od času', color='white', fontsize=11, pad=8)
ax1.legend(fontsize=9, facecolor='#1a1a2e', edgecolor='#555555')
ax1.grid(True, alpha=0.12)
ax1.tick_params(colors='#aaaaaa')

# ── Fázový diagram ────────────────────────────────────────
# Zobrazuje závislosť rýchlosti od výchylky – špirála smeruje do stredu
ax2.set_facecolor('#0d0d1a')
body = ax2.scatter(x, v, c=t, cmap='plasma', s=1.5, alpha=0.8)
ax2.axhline(y=0, color='white', lw=0.8, alpha=0.3)
ax2.axvline(x=0, color='white', lw=0.8, alpha=0.3)
ax2.set_xlabel('Výchylka x (m)', color='#cccccc', fontsize=9)
ax2.set_ylabel('Rýchlosť v (m/s)', color='#cccccc', fontsize=9)
ax2.set_title('Fázový diagram', color='white', fontsize=11, pad=8)
ax2.grid(True, alpha=0.12)
ax2.tick_params(colors='#aaaaaa', labelsize=8)
plt.colorbar(body, ax=ax2, label='Čas (s)')

# ── Graf rýchlosti ────────────────────────────────────────
ax3.set_facecolor('#0d0d1a')
ax3.plot(t, v, color='#FFA500', linewidth=2.0, label='Rýchlosť v(t)')
ax3.axhline(y=0, color='white', linewidth=1, linestyle='--', alpha=0.3)
ax3.fill_between(t, v, 0, where=(v >= 0), alpha=0.08, color='#FFA500')
ax3.fill_between(t, v, 0, where=(v < 0),  alpha=0.08, color='#FF6B6B')
ax3.set_xlabel('Čas t (s)', color='#cccccc', fontsize=10)
ax3.set_ylabel('Rýchlosť v (m/s)', color='#cccccc', fontsize=10)
ax3.set_title('Závislosť rýchlosti od času', color='white', fontsize=11, pad=8)
ax3.legend(fontsize=9, facecolor='#1a1a2e', edgecolor='#555555')
ax3.grid(True, alpha=0.12)
ax3.tick_params(colors='#aaaaaa')

# ── Porovnanie troch režimov ──────────────────────────────
# Vždy zobrazí všetky tri režimy pre aktuálne m a k
ax4.set_facecolor('#0d0d1a')
t_por = np.linspace(0, t_max, 2000)

for b_por, farba, nazov in [
    (b_krit * 0.3, '#00FF88', 'Podtlmený'),   # b = 30% kritického
    (b_krit * 1.0, '#FFD700', 'Kritický'),     # b = presne kritické
    (b_krit * 3.0, '#FF6B6B', 'Pretlmený')    # b = 300% kritického
]:
    sol_por = solve_ivp(
        lambda t, y, bp=b_por: [y[1], (-bp * y[1] - k * y[0]) / m],
        (0, t_max), [x0, v0],
        t_eval=t_por, method='RK45'
    )
    ax4.plot(sol_por.t, sol_por.y[0],
             color=farba, linewidth=1.8,
             linestyle='-' if nazov == 'Kritický' else '--',
             label=nazov, alpha=0.85)

ax4.axhline(y=0, color='white', lw=0.8, alpha=0.3)
ax4.set_xlabel('Čas t (s)', color='#cccccc', fontsize=9)
ax4.set_ylabel('Výchylka x (m)', color='#cccccc', fontsize=9)
ax4.set_title('Porovnanie režimov', color='white', fontsize=11, pad=8)
ax4.legend(fontsize=8.5, facecolor='#1a1a2e', edgecolor='#555555')
ax4.grid(True, alpha=0.12)
ax4.tick_params(colors='#aaaaaa', labelsize=8)

# --- Nadpis celého grafu ---
fig.suptitle(
    f'Tlmený harmonický oscilátor  │  m = {m} kg  │  '
    f'k = {k} N/m  │  b = {b:.1f} N·s/m  │  Režim: {rezim}',
    fontsize=12, color=farba_rezim, y=0.98
)
plt.show()

# --- Výpis vypočítaných hodnôt ---
print(f'{"═"*52}')
print(f'  Hmotnosť:              {m} kg')
print(f'  Tuhosť pružiny:        {k} N/m')
print(f'  Koeficient tlmenia:    {b:.1f} N·s/m')
print(f'{"─"*52}')
print(f'  Vlastná frekvencia:    {omega0:.3f} rad/s  '
      f'({omega0/(2*np.pi):.3f} Hz)')
print(f'  Kritické tlmenie:      {b_krit:.2f} N·s/m')
print(f'  Pomer b/b_krit:        {pomer:.3f}')
print(f'{"─"*52}')
print(f'  Režim tlmenia:         {rezim}')
print(f'{"═"*52}')
```
