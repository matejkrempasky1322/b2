```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
%matplotlib inline
plt.style.use('dark_background')

# ============================================================
# AKO POUŽÍVAŤ TENTO NOTEBOOK V JUPYTERLAB:
# 1. Zmeň hodnoty v bloku PARAMETRE nižšie
# 2. Spusti bunku klávesou Shift+Enter
# 3. Alebo použi slidery ktoré sa zobrazia pod bunkou
# ============================================================

# ============================================================
# PARAMETRE – môžeš meniť tu ALEBO cez slidery nižšie
# ============================================================
v0_kmh   = 50     # počiatočná rýchlosť [km/h]   → 0 až 200
a        = -3.0   # zrýchlenie [m/s²]             → záporné = brzdenie, kladné = zrýchlenie
t_max    = 10     # čas simulácie [s]             → 2 až 30
jednotky = 'm'    # jednotky polohy               → 'm' alebo 'km'
# ============================================================

def zobraz_kinematiku(v0_kmh, a, t_max, jednotky):
    v0    = v0_kmh / 3.6          # prepočet km/h → m/s
    t     = np.linspace(0, t_max, 500)

    # Výpočet kinematických veličín
    s     = v0 * t + 0.5 * a * t**2   # poloha [m]
    v     = v0 + a * t                  # rýchlosť [m/s]
    v_kmh = v * 3.6                     # rýchlosť [km/h]
    acc   = np.full_like(t, a)          # zrýchlenie – konštanta

    # Odrežeme čas keď vozidlo zastalo (rýchlosť < 0)
    mask  = v >= 0
    t     = t[mask]
    s     = s[mask]
    v_kmh = v_kmh[mask]
    acc   = acc[mask]

    # Prepočet jednotiek polohy
    if jednotky == 'km':
        s_plot  = s / 1000
        s_label = 'Poloha s (km)'
    else:
        s_plot  = s
        s_label = 'Poloha s (m)'

    fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 8), sharex=True)
    fig.suptitle(f'Kinematika pohybu  |  v₀ = {v0_kmh} km/h  |  a = {a} m/s²',
                 fontsize=13, color='white')

    ax1.plot(t, s_plot, color='#00BFFF', linewidth=2)
    ax1.set_ylabel(s_label, color='white')
    ax1.set_title('Závislosť polohy od času', color='#aaaaaa', fontsize=10)
    ax1.grid(True, alpha=0.2)
    ax1.fill_between(t, s_plot, alpha=0.15, color='#00BFFF')

    ax2.plot(t, v_kmh, color='#FFA500', linewidth=2)
    ax2.set_ylabel('Rýchlosť v (km/h)', color='white')
    ax2.set_title('Závislosť rýchlosti od času', color='#aaaaaa', fontsize=10)
    ax2.grid(True, alpha=0.2)
    ax2.fill_between(t, v_kmh, alpha=0.15, color='#FFA500')

    ax3.plot(t, acc, color='#90EE90', linewidth=2)
    ax3.set_ylabel('Zrýchlenie a (m/s²)', color='white')
    ax3.set_xlabel('Čas t (s)', color='white')
    ax3.set_title('Závislosť zrýchlenia od času', color='#aaaaaa', fontsize=10)
    ax3.grid(True, alpha=0.2)
    ax3.fill_between(t, acc, alpha=0.15, color='#90EE90')

    plt.tight_layout()
    plt.show()

_ = widgets.interact(
    zobraz_kinematiku,
    v0_kmh=widgets.FloatSlider(
        value=v0_kmh, min=0, max=200, step=5,
        description='v₀ (km/h):',
        style={'description_width': '120px'},
        layout=widgets.Layout(width='500px')),
    a=widgets.FloatSlider(
        value=a, min=-10, max=5, step=0.5,
        description='a (m/s²):',
        style={'description_width': '120px'},
        layout=widgets.Layout(width='500px')),
    t_max=widgets.FloatSlider(
        value=t_max, min=2, max=30, step=1,
        description='Čas (s):',
        style={'description_width': '120px'},
        layout=widgets.Layout(width='500px')),
    jednotky=widgets.RadioButtons(
        options=['m', 'km'],
        value=jednotky,
        description='Jednotky:',
        style={'description_width': '120px'})
)
```
