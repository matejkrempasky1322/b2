# Jupyter

Platforma [Jupyter](https://jupyter.org/) je primárne určená na vytváranie interaktívnych textov (notebooky), kde je možné kombinovať bežný text formátovaný pomocou Markdown, matematické výrazy a vykonateľné kódy v jazyku Python, s pomocou rozšírení aj ďalších programovacích jazykoch. Platforma využíva prostredie www prehliadača a pre jej beh je potrebné nainštalovať lokálny server (v prípade použitia na domácom počítači).

## Inštalácia

Inštalácia serveru JupyterLab z [Python Package Index](https://pypi.org/project/Sphinx/) (PyPI) pomocou inštalátora pip:

```python
# Inštalácia JupyterLab
pip install jupyterlab
```

Inštalácia modifikácií prostredia Jupyter ako aj inštalácia pre rôzne platformy je uvedená v [dokumentácii](https://jupyter.org/install). Po inštalácii je vhodné vytvoriť zdrojový adresár pre budúce stránky notebookov a v tomto adresári spustiť príkazom v konzole server:

```python
#Spustenie JupyterLab
python -m jupyterlab
```

Po spustení sa v predvolenom www prehliadači otvorí aplikácia JupyterLab, v ktorej môžeme vytvárať stránky notebookov.
