# Sphinx

Sphinx je platforma pre konvertovanie obyčajného textu do rôznych výstupných formátov, pričom vytvára indexy, krížové odkazy, vkladá obrázky, formátuje text, matematické vzťahy a podporuje mnohé ďalšie nástroje na tvorbu dokumentácie, blogov alebo kníh. Zdrojové texty sú uložené v textových súboroch vo formátoch reStructuredText alebo Markdown.

Súčasťou platformy je množstvo predlôh generovaných testov ako aj rozšírení a doplnkov na ich úpravy.

## Markdown vs reStructuredText

Vytváranie textov vo formáte Markdown má jednoduchú syntax, formát je vhodný na vytváranie dokumentov tvorených manuálne, na druhej strane neposkytuje všetky možnosti úpravy textov ako reStructuredText.

reStructuredText má širšie možnosti vytvárania a formátovania tabuliek, automatické generovanie obsahu, integrácia so systémami programového generovania dokumentácie a je vhodný na tvorbu dokumentov vytváraných strojovo.

## Inštalácia

Inštalácia Sphinx z PyPI pomocou pip:

```python
pip install sphinx sphinx-rtd-theme nbsphinx
```

## Spustenie

Po inštalácii vytvoríme konfiguračný súbor príkazom:

```python
sphinx-quickstart
```

Príkaz nás prevedie nastavením projektu. Po dokončení zostavíme stránku:

```python
sphinx-build -b html . _build/html
```
