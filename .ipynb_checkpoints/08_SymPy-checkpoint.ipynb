# Symbolická matematika s knižnicou SymPy

SymPy je Python knižnica pre symbolickú matematiku – pracujeme
so symbolmi a výrazmi namiesto čísel, výsledky sú analytické vzorce.

V tomto notebooku ukážeme hlavné funkcionality SymPy:
základnú aritmetiku, derivovanie, integrovanie, riešenie rovníc
a diferenciálnych rovníc, limity, Taylorov rad a lineárnu algebru.

## Základná symbolická aritmetika

```python
import sympy as sp
from IPython.display import display

# Definícia symbolov
x, y, z, n = sp.symbols('x y z n')
a, b, c    = sp.symbols('a b c', positive=True)

# Racionálne čísla – exaktne, bez zaokrúhlenia
print('Racionálne čísla:')
display(sp.Rational(1, 3) + sp.Rational(1, 6))   # = 1/2

# Symbolická mocnina a koreň
print('Symbolické výrazy:')
display(sp.sqrt(8))                               # = 2√2
display(sp.sqrt(x**2))                            # = |x|

# Rozvoj výrazu
vyraz = (x + y)**4
print('Rozvoj (x+y)⁴:')
display(sp.expand(vyraz))

# Faktorizácia
polynom = x**3 - x**2 - x + 1
print('Faktorizácia x³ - x² - x + 1:')
display(sp.factor(polynom))

# Zjednodušenie
zlomok = (x**2 - 1) / (x - 1)
print('Zjednodušenie (x²-1)/(x-1):')
display(sp.simplify(zlomok))

# Trigonometrická identita
print('Trigonometrická identita:')
display(sp.trigsimp(sp.sin(x)**2 + sp.cos(x)**2))  # = 1
```

## Derivovanie – sp.diff()

```python
import sympy as sp
from IPython.display import display

x, t = sp.symbols('x t')
v0, a_sym, k, m = sp.symbols('v_0 a k m', positive=True)

# Jednoduchá derivácia
f = x**4 - 3*x**2 + 2*x - 5
print('f(x) =')
display(f)
print("f'(x) =")
display(sp.diff(f, x))
print("f''(x) =")
display(sp.diff(f, x, 2))

# Kinematika – poloha → rýchlosť → zrýchlenie
s = v0*t - sp.Rational(1, 2)*a_sym*t**2
print('\nPoloha pri brzdení s(t):')
display(s)
print('Rýchlosť v(t) = ds/dt:')
display(sp.diff(s, t))
print('Zrýchlenie a(t) = dv/dt:')
display(sp.diff(s, t, 2))

# Parciálna derivácia
f2 = x**2 * t + sp.sin(x*t)
print('\nf(x,t) = x²t + sin(xt)')
print('∂f/∂x =')
display(sp.diff(f2, x))
print('∂f/∂t =')
display(sp.diff(f2, t))
```

## Integrovanie – sp.integrate()

```python
import sympy as sp
from IPython.display import display

x, t = sp.symbols('x t')
v0, a_sym = sp.symbols('v_0 a', positive=True)

# Neurčitý integrál
print('Neurčitý integrál x³:')
display(sp.integrate(x**3, x))

print('Neurčitý integrál sin²(x):')
display(sp.integrate(sp.sin(x)**2, x))

print('Neurčitý integrál e^x·sin(x):')
display(sp.integrate(sp.exp(x)*sp.sin(x), x))

# Určitý integrál
print('\nUrčitý integrál ∫₀^π sin(x) dx:')
display(sp.integrate(sp.sin(x), (x, 0, sp.pi)))

print('Plocha pod Gaussovkou ∫₋∞^∞ e^(-x²) dx:')
display(sp.integrate(sp.exp(-x**2), (x, -sp.oo, sp.oo)))

# Fyzikálna aplikácia: brzdná dráha = ∫v dt
v = v0 - a_sym*t
t_stop = v0 / a_sym
print('\nBrzdná dráha = ∫₀^t_stop v dt:')
display(sp.integrate(v, (t, 0, t_stop)))
```

## Riešenie rovníc – sp.solve() a sp.dsolve()

```python
import sympy as sp
from IPython.display import display

x, t = sp.symbols('x t')
v0, a_sym, m_sym, k_sym, b_sym = sp.symbols('v_0 a m k b', positive=True)

# Algebraické rovnice
print('Riešenie x² - 5x + 6 = 0:')
display(sp.solve(x**2 - 5*x + 6, x))

print('\nRiešenie x³ - 6x² + 11x - 6 = 0:')
display(sp.solve(x**3 - 6*x**2 + 11*x - 6, x))

# Sústava rovníc
print('\nSústava: x + y = 5, x - y = 1:')
y = sp.Symbol('y')
display(sp.solve([x + y - 5, x - y - 1], [x, y]))

# Riešenie v=0 → čas zastavenia (kinematika)
v = v0 - a_sym*t
print('\nČas zastavenia (v=0):')
display(sp.solve(v, t))

# Diferenciálna rovnica – tlmený oscilátor
# m·x'' + b·x' + k·x = 0
x_func = sp.Function('x')
ode = m_sym*x_func(t).diff(t, 2) + b_sym*x_func(t).diff(t) + k_sym*x_func(t)
print('\nAnalytické riešenie tlmeného oscilátora m·x\'\'+b·x\'+k·x=0:')
display(sp.dsolve(ode, x_func(t)))
```

## Limity a Taylorov rad – sp.limit(), sp.series()

```python
import sympy as sp
from IPython.display import display

x, h = sp.symbols('x h')

# Limity
print('lim(x→0) sin(x)/x:')
display(sp.limit(sp.sin(x)/x, x, 0))

print('\nlim(x→∞) (1 + 1/x)^x:')
display(sp.limit((1 + 1/x)**x, x, sp.oo))

print('\nlim(x→0) (e^x - 1)/x:')
display(sp.limit((sp.exp(x) - 1)/x, x, 0))

# Taylorov rad
print('\nTaylorov rad sin(x) okolo x=0 (rád 7):')
display(sp.series(sp.sin(x), x, 0, 7))

print('\nTaylorov rad e^x okolo x=0 (rád 6):')
display(sp.series(sp.exp(x), x, 0, 6))

print('\nTaylorov rad cos(x) okolo x=0 (rád 6):')
display(sp.series(sp.cos(x), x, 0, 6))
```

## Lineárna algebra – sp.Matrix()

```python
import sympy as sp
from IPython.display import display

# Definícia matíc
A = sp.Matrix([[1, 2], [3, 4]])
B = sp.Matrix([[5, 6], [7, 8]])

print('Matica A:')
display(A)

print('Determinant A:')
display(sp.det(A))

print('Inverzná matica A⁻¹:')
display(A.inv())

# Symbolická matica – sústava rovníc
x, y = sp.symbols('x y')
M = sp.Matrix([[2, 3], [1, -1]])
b_vec = sp.Matrix([7, 1])
print('\nRiešenie sústavy M·[x,y] = b:')
display(M.solve(b_vec))

# Matica rotácie
theta = sp.Symbol('theta')
R = sp.Matrix([
    [sp.cos(theta), -sp.sin(theta)],
    [sp.sin(theta),  sp.cos(theta)]
])
print('\nMatica rotácie R(θ):')
display(R)
print('R·Rᵀ = I:')
display(sp.simplify(R * R.T))
```

## LaTeX a lambdify – sp.latex(), sp.lambdify()

```python
import sympy as sp
import numpy as np
from IPython.display import display, Math

x, t = sp.symbols('x t')

# LaTeX výstup – automatické generovanie kódu
vyraz = sp.Integral(sp.sin(x)**2 / sp.sqrt(1 - x**2), x)
print('LaTeX kód výrazu:')
print(sp.latex(vyraz))
print()
display(Math(sp.latex(vyraz)))

# Komplexnejší vzorec
f = sp.Rational(1, 2) * sp.sqrt(sp.pi) * sp.erf(x)
print('\nLaTeX pre 1/2·√π·erf(x):')
print(sp.latex(f))
display(Math(sp.latex(f)))

# Lambdify – symbolický výraz → rýchla numerická funkcia
print('\nLambdify – konverzia na numerickú funkciu:')
s_sym = sp.Symbol('v_0') * t - sp.Rational(1, 2) * sp.Symbol('a') * t**2
s_sub = s_sym.subs([(sp.Symbol('v_0'), 22.2), (sp.Symbol('a'), 5.0)])
s_num = sp.lambdify(t, s_sub, 'numpy')   # numpy backend = rýchle pole

t_arr = np.linspace(0, 4.44, 5)
print(f'Poloha pri brzdení 80 km/h, a=5 m/s²:')
for ti, si in zip(t_arr, s_num(t_arr)):
    print(f'  t = {ti:.2f} s  →  s = {si:.2f} m')
```

## Prehľad funkcií SymPy

| Kategória | Funkcia | Popis |
|-----------|---------|-------|
| Základy | `sp.symbols()`, `sp.Rational()` | Definícia symbolov a exaktných čísel |
| Úprava | `sp.expand()`, `sp.factor()`, `sp.simplify()` | Rozvoj, faktorizácia, zjednodušenie |
| Derivovanie | `sp.diff(f, x, n)` | n-tá derivácia f podľa x |
| Integrovanie | `sp.integrate(f, x)` | Neurčitý integrál |
| | `sp.integrate(f, (x, a, b))` | Určitý integrál |
| Rovnice | `sp.solve(f, x)` | Algebraické rovnice |
| | `sp.dsolve(ode)` | Diferenciálne rovnice |
| Limity | `sp.limit(f, x, x0)` | Limita funkcie |
| Rady | `sp.series(f, x, x0, n)` | Taylorov rad rádu n |
| Algebra | `sp.Matrix()`, `.det()`, `.inv()`, `.eigenvects()` | Maticová algebra |
| Výstup | `sp.latex(f)` | LaTeX kód výrazu |
| | `sp.lambdify(x, f)` | Konverzia na numerickú funkciu |
