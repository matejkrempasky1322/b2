# Matematické výrazy

Pre písanie matematických výrazov je využívaný typografický systém [LaTeX](https://en.wikibooks.org/wiki/LaTeX/Mathematics). Matematické výrazy je možné do textu dokumentu zaradiť niekoľkými spôsobmi.

## Príkaz {math}

Príkaz `{math}` je určený pre písanie jednoduchých rovníc alebo sústav rovníc. Je ekvivalentom príkazu `{align}` v systéme LaTeX. Rovnice sú oddelené znakmi `\\` a vzájomne zarovnané znakom `&`.

V texte môžeme zadať matematický výraz vo formáte `` {math}`vyraz` ``.

```
{math}
:label: referencia
   equ_1 \\
   equ_2 \\
   ...
```

Na sústavu rovníc sa môžeme odkazovať pomocou referencie v tvare `` {eq}`referencia` ``.

### Príklad 1. Použitie príkazu {math}

```
Vzťah medzi amplitúdou $A$, fázou $\phi$ a zložkami $I$ a $Q$:

$$
A = \sqrt{I^2 + Q^2}
$$

$$
\phi = \arctan{\frac{Q}{I}}
$$
```

## Rozšírenie dollarmath

Rozšírenie `dollarmath` umožňuje jednoduché zadávanie matematických výrazov v texte pomocou reťazca `$výraz$` a sústavy rovníc vzájomne zarovnaných znakom `&`:

```
$$
   equ_1 \\
   equ_2 \\
   ...
$$(referencia)
```

Príklad použitia dollarmath:

$A = \sqrt{I^2 + Q^2}$

$$
A &= \sqrt{I^2 + Q^2} \\
\phi &= \arctan{\frac{Q}{I}}
$$(amp_phase2)

## Rozšírenie amsmath

Rozšírenie [amsmath](https://www.sphinx-doc.org/) umožňuje zadávať matematické výrazy v syntaxi systému LaTeX s využitím jeho prostredí `equation`, `align`, `matrix` a ďalších.

````python
\begin{equation*} 
f(a)+\frac{f'(a)}{1!}(x-a) + 
\frac{f''(a)}{2!}(x-a)^2 + 
\frac{f^{(3)}(a)}{3!}(x-a)^3 + \cdots
\end{equation*}
````

Celý výraz môžeme zobraziť farebne použitím HTML značky `<span>`:

<span style='color:#2ecc71'>

$$
q = \sqrt{\sin(\alpha^2) + \cos(\beta^2)}
$$

</span>
