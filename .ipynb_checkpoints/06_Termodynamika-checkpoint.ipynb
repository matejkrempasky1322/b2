# Termodynamika spaľovacieho motora

Spaľovací motor je tepelný stroj, ktorý premieňa tepelnú energiu
paliva na mechanickú prácu. Účinnosť tohto procesu je obmedzená
základnými zákonmi termodynamiky. V tomto notebooku skúmame
Carnotov cyklus ako ideálny model tepelného stroja.

Pri otvorení v JupyterLab sú k dispozícii interaktívne slidery –
hodnoty môžeš meniť aj priamo v bloku PARAMETRE v kóde.

## Teoretický základ

**Carnotov cyklus** pozostáva zo štyroch dejov:

1. **Izotermická expanzia** pri teplote $T_H$ – plyn prijíma teplo $Q_H$
2. **Adiabatická expanzia** – plyn sa ochladzuje z $T_H$ na $T_C$
3. **Izotermická kompresia** pri teplote $T_C$ – plyn odovzdáva teplo $Q_C$
4. **Adiabatická kompresia** – plyn sa ohrieva z $T_C$ na $T_H$

**Tepelná účinnosť** Carnotovho cyklu:
$$\eta = 1 - \frac{T_C}{T_H}$$

kde $T_H$ je teplota teplého zásobníka a $T_C$ je teplota studeného
zásobníka, obe v Kelvinoch.

**Prepočet teplôt:**
$$T[K] = T[°C] + 273{,}15$$

## p-V diagram a výpočet účinnosti

![Carnotov cyklus](_images/termodynamika_1.png)
*Obr. Carnotov cyklus – p-V diagram a závislosť účinnosti od teploty*

```python
# ============================================================
# AKO POUŽÍVAŤ TENTO NOTEBOOK V JUPYTERLAB:
# 1. Zmeň hodnoty v bloku PARAMETRE nižšie
# 2. Spusti bunku klávesou Shift+Enter
# 3. Alebo použi slidery ktoré sa zobrazia pod bunkou
# ============================================================

import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output
%matplotlib inline
plt.style.use('dark_background')

# ============================================================
# PARAMETRE – môžeš meniť tu ALEBO cez slidery nižšie
# ============================================================
T_H_C_def = 400    # teplota teplého zásobníka [°C]    → spaľovanie: 1800–2200 °C
T_C_C_def = 50     # teplota studeného zásobníka [°C]  → výfuk: 100–150 °C
# ============================================================

# Fyzikálne konštanty pre p-V diagram – nemeniť
n     = 1      # množstvo plynu [mol]
R     = 8.314  # plynová konštanta [J/(mol·K)]
gamma = 1.4    # Poissonov koeficient (dvojatómový plyn)
V1    = 1.0    # počiatočný objem [×10⁻³ m³]
V2    = 3.0    # objem po izotermickej expanzii [×10⁻³ m³]

out = widgets.Output()

def zobraz_carnot(T_H_C, T_C_C):
    # Prepočet teplôt z °C na K
    T_H = T_H_C + 273.15
    T_C = T_C_C + 273.15

    # Carnotova účinnosť: η = 1 - T_C/T_H
    eta = 1 - T_C / T_H

    # Výpočet stavových bodov cyklu
    V_1 = V1
    p_1 = n * R * T_H / V_1              # bod 1: začiatok izotermickej expanzie
    V_2 = V2
    p_2 = n * R * T_H / V_2              # bod 2: koniec izotermickej expanzie
    V_3 = V_2 * (T_H / T_C) ** (1 / (gamma - 1))   # bod 3: koniec adiabatickej expanzie
    p_3 = n * R * T_C / V_3
    V_4 = V_1 * (T_H / T_C) ** (1 / (gamma - 1))   # bod 4: koniec izotermickej kompresie
    p_4 = n * R * T_C / V_4

    # Výpočet vetiev cyklu (200 bodov na každej vetve)
    V_12 = np.linspace(V_1, V_2, 200)
    p_12 = n * R * T_H / V_12            # izotermická expanzia (konštantná T_H)

    V_23 = np.linspace(V_2, V_3, 200)
    p_23 = p_2 * (V_2 / V_23) ** gamma  # adiabatická expanzia (p·V^γ = konšt.)

    V_34 = np.linspace(V_3, V_4, 200)
    p_34 = n * R * T_C / V_34            # izotermická kompresia (konštantná T_C)

    V_41 = np.linspace(V_4, V_1, 200)
    p_41 = p_4 * (V_4 / V_41) ** gamma  # adiabatická kompresia

    # Definícia farieb a popisov vetiev
    vetvy = [
        (V_12, p_12, '#00BFFF', '1→2  Izotermická expanzia ($T_H$)'),
        (V_23, p_23, '#00FF88', '2→3  Adiabatická expanzia'),
        (V_34, p_34, '#FF6B6B', '3→4  Izotermická kompresia ($T_C$)'),
        (V_41, p_41, '#FFA500', '4→1  Adiabatická kompresia'),
    ]

    with out:
        clear_output(wait=True)

        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))
        fig.patch.set_facecolor('#0d0d1a')

        # ── p-V diagram ───────────────────────────────────────
        ax1.set_facecolor('#0d0d1a')
        for V, p, farba, label in vetvy:
            ax1.plot(V, p, color=farba, linewidth=2.5, label=label)

        # Označenie stavových bodov
        for i, (vb, pb) in enumerate(zip(
                [V_1, V_2, V_3, V_4],
                [p_1, p_2, p_3, p_4])):
            ax1.plot(vb, pb, 'o', color='white', markersize=8, zorder=5)
            ax1.text(vb + 0.05, pb + 0.02, f'  {i+1}',
                     color='white', fontsize=11, fontweight='bold')

        # Vyplnenie plochy cyklu (žltá = vykonaná práca)
        V_uzavr = np.concatenate([v[0] for v in vetvy])
        p_uzavr = np.concatenate([v[1] for v in vetvy])
        ax1.fill(V_uzavr, p_uzavr, alpha=0.12, color='#FFD700')

        ax1.set_xlabel('Objem V (×10⁻³ m³)', color='#cccccc')
        ax1.set_ylabel('Tlak p (kPa)', color='#cccccc')
        ax1.set_title('Carnotov cyklus – p-V diagram', color='white', fontsize=11)
        ax1.legend(fontsize=8.5, facecolor='#1a1a2e', edgecolor='#555555')
        ax1.grid(True, alpha=0.12)
        ax1.tick_params(colors='#aaaaaa')
        ax1.text(0.02, 0.97, f'$T_H$ = {T_H_C} °C  ({T_H:.0f} K)',
                 transform=ax1.transAxes, color='#00BFFF', fontsize=9, va='top')
        ax1.text(0.02, 0.91, f'$T_C$ = {T_C_C} °C  ({T_C:.0f} K)',
                 transform=ax1.transAxes, color='#FF6B6B', fontsize=9, va='top')

        # ── Graf závislosti účinnosti od teploty ──────────────
        ax2.set_facecolor('#0d0d1a')

        # Rozsah teplôt pre krivku účinnosti
        T_H_range = np.linspace(T_C + 10, T_C + 1500, 300)
        eta_range = 1 - T_C / T_H_range   # Carnotova účinnosť pre každú T_H

        ax2.plot(T_H_range - 273.15, eta_range * 100,
                 color='#00BFFF', linewidth=2.5, label='Carnotova účinnosť')
        ax2.axvline(x=T_H_C, color='#FFD700', linewidth=2,
                    linestyle='--', alpha=0.9,
                    label=f'Aktuálne $T_H$ = {T_H_C} °C')
        ax2.plot(T_H_C, eta * 100, 'o', color='#FFD700', markersize=10, zorder=5)
        ax2.annotate(f'η = {eta*100:.1f} %',
                     xy=(T_H_C, eta * 100),
                     xytext=(T_H_C + 50, eta * 100 - 5),
                     color='#FFD700', fontsize=11, fontweight='bold',
                     arrowprops=dict(arrowstyle='->', color='#FFD700', lw=1.5))
        ax2.fill_between(T_H_range - 273.15, eta_range * 100,
                         alpha=0.10, color='#00BFFF')

        ax2.set_xlabel('Teplota teplého zásobníka $T_H$ (°C)', color='#cccccc')
        ax2.set_ylabel('Účinnosť η (%)', color='#cccccc')
        ax2.set_title('Závislosť účinnosti od teploty', color='white', fontsize=11)
        ax2.legend(fontsize=9, facecolor='#1a1a2e', edgecolor='#555555')
        ax2.grid(True, alpha=0.12)
        ax2.tick_params(colors='#aaaaaa')
        ax2.set_ylim(0, 100)

        fig.suptitle(
            f'Carnotov cyklus  │  $T_H$ = {T_H_C} °C  │  '
            f'$T_C$ = {T_C_C} °C  │  η = {eta*100:.1f} %',
            fontsize=12, color='white'
        )
        plt.tight_layout()
        plt.show()

        # Výpis vypočítaných hodnôt
        print(f'Teplota teplého zásobníka:    {T_H_C} °C  ({T_H:.1f} K)')
        print(f'Teplota studeného zásobníka:  {T_C_C} °C  ({T_C:.1f} K)')
        print(f'Carnotova účinnosť:           {eta*100:.2f} %')

_ = widgets.interact(
    zobraz_carnot,
    T_H_C=widgets.IntSlider(
        value=T_H_C_def, min=100, max=2200, step=10,
        description='T_H (°C):',
        style={'description_width': '150px'},
        layout=widgets.Layout(width='500px')),
    T_C_C=widgets.IntSlider(
        value=T_C_C_def, min=0, max=500, step=5,
        description='T_C (°C):',
        style={'description_width': '150px'},
        layout=widgets.Layout(width='500px'))
)
display(out)
```

## Porovnanie s reálnym spaľovacím motorom

Typický spaľovací motor pracuje pri teplotách:
- Teplota spaľovania: $T_H \approx 1800\text{–}2200$ °C
- Teplota výfuku: $T_C \approx 100\text{–}150$ °C

Reálne motory dosahujú len $30\text{–}40$ % z teoretickej maximálnej
účinnosti kvôli treniu, tepelným stratám a nedokonalému spaľovaniu.
