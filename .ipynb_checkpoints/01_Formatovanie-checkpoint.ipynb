# Formátovanie textu

Zdrojový text dokumentu vytvárame v ľubovoľnom textovom editore pre editovanie natívneho textu. Je vhodné aby editor rozpoznával a farebne zvýrazňoval syntax formátovania Markdown, napríklad [Geany](https://www.geany.org/), [Visual Studio Code](https://code.visualstudio.com/), [Kate](https://kate-editor.org/).

## Komentáre

Znakom `%` označíme v zdrojovom texte dokumentu riadky s komentárom, poznámkami alebo časti dokumentu ktoré budú vyradené z generovania výsledného dokumentu:

```python
% komentár v dokumente je pri konverzii textu ignorovaný
```

## Záhlavie dokumentu

Každá samostatná kapitola dokumentu musí začínať záhlavím začínajúcim jedným znakom mriežky `#`, označenia podkapitol začínajú opakovaním tohoto znaku až do 6. úrovne. Záhlavia v hlavnom texte sa prenášajú do obsahu dokumentu.

```python
# Názov kapitoly
   ## Podkapitola 1.
      ### Odsek 1.
      ### Odsek 2.
   ## Podkapitola 2.
```

# Názov kapitoly
   ## Podkapitola 1.
      ### Odsek 1.
      ### Odsek 2.
   ## Podkapitola 2.

> **Poznámka:** Konvertor Sphinx kontroluje hierarchické usporiadanie textu, pri vynechaní názvu kapitoly alebo niektorej úrovne podkapitol vypíše pri konverzii upozornenie.

## Typ textu

Základné formátovanie textu získame použitím hviezdičkovej konvencie:

```python
*šikmý text*
**tučný text**
***šikmý tučný text***
```

*šikmý text*
**tučný text**
***šikmý tučný text***

> **Poznámka:** Text musí byť medzi hviezdičkami uzatvorený bez medzier na začiatku a konci, v opačnom prípade je formátovanie ignorované.

## Farba textu

Farebnú reprezentáciu textu môžeme definovať pomocou [HTML značiek](https://www.w3schools.com/html/default.asp). Pre zmenu farby textu môžeme použiť značku `<font>`:

```python
<font color='COLOR_NAME'> text </font>
<font color='#RRGGBB'> text </font>
```

<font color='COLOR_NAME'> text </font>
<font color='#RRGGBB'> text </font>

kde `COLOR_NAME` je meno pomenovanej farby. Kompletný zoznam mien farieb rozpoznávaných www prehliadačmi je uvedený v [HTML dokumentácii](https://www.w3schools.com/colors/colors_names.asp).

Farbu textu môžeme zadať aj pomocou hexadecimálneho číselného kódu vo formáte `#RRGGBB`. Prevod zvolenej farby na jej kód je súčasťou každého grafického editora, interaktívne si môžeme vygenerovať kódy na [internete](https://htmlcolorcodes.com/).

Širšie možnosti formátovania farebného zobrazenia textu umožňuje atribút `style`:

```python
<span style='color: COLOR_NAME;'> text </span>
<p    style='color: #RRGGBB; background-color: #RRGGBB;'> text </p>
<h1   style='border: 2px solid red; border-radius: 5px;'> text </h1>
```

<span style='color: COLOR_NAME;'> text </span>
<p    style='color: #RRGGBB; background-color: #RRGGBB;'> text </p>
<h1   style='border: 2px solid red; border-radius: 5px;'> text </h1>

## Príklady farebného formátovania

Príklad 1. Zvýraznenie inline textu:

```python
<span style='color: #2ecc71; font-weight: bold;'>Zvýraznený text zelenou farbou</span>
```

<span style='color: #2ecc71; font-weight: bold;'>Zvýraznený text zelenou farbou</span>

Príklad 2. Zvýraznenie paragrafu:

```python
<p style='background-color: #2c3e50; color: white; padding: 10px; border-radius: 5px;'>
Zvýraznený paragraf s tmavým pozadím.
</p>
```

<p style='background-color: #2c3e50; color: white; padding: 10px; border-radius: 5px;'>
Zvýraznený paragraf s tmavým pozadím.
</p>

> **Poznámka:** HTML značka `<font>` je vo verzii HTML 5 pokladaná za zastaralú a jej používanie sa v nových dokumentoch neodporúča. Odporúča sa používať atribút `style`.
